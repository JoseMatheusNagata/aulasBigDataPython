```python
import pandas as pd

data= {
    'nome': ['alice', 'bob', 'clara'],
    'idade': [25, 30, 35],
    'cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']
}

#conversao do dicionario em dataFrame
df= pd.DataFrame(data)

#Exibicao do DataFrame completo
print(df)

#selecao de uma ou mais colunas
print("\nseleciona somente a coluna nome")
print(df['nome'])

#selecionar multiplas colunas
print("\nseleciona as colunas nome e idade")
print(df[['nome', 'idade']])

#filtrar um novo dataFrame que comtem appenas as linhas com pessoas com mais de 28 anos
print("pessoas com mais de 28 anos")
df_maior_28= df[df['idade']>28]
print(df_maior_28)

#adicao de uma nova colula salario
df['salario']= [5000, 6000, 7000]
print("\nDataFrame com a nova coluna salario")
print(df)

```

```python
import pandas as pd

#seride com dados de temperatura em graus celsius
temperaturas = pd.Series([25, 28, 22, 29, 30, 24, 38])

print("serie de temperaturas: ")
print(temperaturas)

#Calculando a media tem temperaturas
media_temperaturas = temperaturas.mean()
print(f"Media das temperaturas: {media_temperaturas:.2f}C")

#contando a quantidade de valores
contagem_valores = temperaturas.count()
print(f"Quantidade de valores: {contagem_valores}")

#Descrevendo estatisticas basicas da Series
print("\nEstatisticas da serie de temperaturas:")
print(temperaturas.describe())

#Arredondando os valores da seria
print("\nValores arredondados:")
print(temperaturas.round(2))
print(temperaturas)
```

```python
import pandas as pd

#criando uma Series com indice personalizado
#Vamos representar uma populacao de algumas cidades
populacao_cidades = pd.Series([280000,1500000,1000000,500000], index=['Sao Paulo','Rio de Janeiro','Belo Horizonte','Porto Alegre'])
print(populacao_cidades)

#Acessando um elemento pelo seu rotulo(nome da cidade)
print("\nPopulacao de Sao Paulo: ", populacao_cidades['Sao Paulo'])

#Realizando operacoes matematicas simples
print("\nPopulacao total das 4 cidades: ", populacao_cidades.sum())

#filtrando dados com base em uma condicao
cidades_acima_de_1milhao = populacao_cidades[populacao_cidades > 1000000]
print("Cidade com populacao maior que 1 milhao: ")
print(cidades_acima_de_1milhao)
```

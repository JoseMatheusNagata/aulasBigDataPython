```python
import pandas as pd

#criando uma Series a partir de uma lista de dados
#Vamos representar uma populacao de algumas cidades
populacao_cidades = pd.Series([280000,1500000,1000000,500000])

print("Series de população de cidades: ")
print(populacao_cidades)

#Acessando um elemento pelo indice numero padrao
print("\nPopulaçãodo primeiro elemento: ", populacao_cidades[0])
print("População do ultimo elemento: ", populacao_cidades[3])
```

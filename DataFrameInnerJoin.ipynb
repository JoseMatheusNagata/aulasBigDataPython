```python
import pandas as pd

#dataFrame de vendas
data_vendas= {
    'id_venda': [1,2,3,4],
    'id_produto' : [101,102,101,103],
    'quantidade' : [2,1,3,2],
}
df_vendas = pd.DataFrame(data_vendas)

#dataFrame de produtos
data_produtos={
    'id_produto' : [101,102,103],
    'nome_produto' : ['Camiseta','Calça','Tênis'],
    'preco' : [29.99,39.99,49.99],
}
df_produtos = pd.DataFrame(data_produtos)

df_merge_inner = pd.merge(df_vendas,df_produtos,on='id_produto',how='inner')

# Calcula o total vendido por linha
df_merge_inner['total_vendido'] = df_merge_inner['quantidade'] * df_merge_inner['preco']

print(df_merge_inner)
```

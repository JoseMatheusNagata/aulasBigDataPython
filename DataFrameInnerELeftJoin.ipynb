```python
import pandas as pd

#DataFrame dos pedidos
pedidos = pd.DataFrame({
    'id_pedido': [101,102,103,104],
    'id_cliente': [1,2,1,3],
    'produto': ['Celular','TC','Tablet','Fone']
})

#DataFrame de clientes
clientes = pd.DataFrame({
    'id_cliente': [1,2,4],
    'nome_cliente': ['Ana','Bruno','Carlos']
})

print("pedidos:")
print(pedidos)
print("\n")
print("clientes:")
print(clientes)

df_merge_inner = pd.merge(pedidos,clientes,on='id_cliente',how='inner')
print("\n")
print("pedidos e clientes:")
print(df_merge_inner)

df_merge_left = pd.merge(pedidos,clientes,on='id_cliente',how='left')
print("\n")
print("pedidos e clientes:")
print(df_merge_left)
```

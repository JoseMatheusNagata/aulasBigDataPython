```python
import pandas as pd

#Crie dados em um dicionario de produtos com as colulas: produdo, preço, quantidade de estoque, disponivel(True ou false)
dados_produtos= {
    'Produto': ['Celular', 'NoteBook', 'Mouse', 'Teclado'],
    'Preco(R$)': [1500, 4200, 80, 150],
    'Quantidade em Estoque': [50, 20, 150, 100],
    'Disponivel': [True, True, True, False]

}

#Criando o dataFrame
df_produtos = pd.DataFrame(dados_produtos)

#Exibindo o dataFrame
print(df_produtos)

#analise e manupulacao basica
#acessando uma colula
print("acessando a produto: ")
print(df_produtos['Produto'])

#filtrando linha com base em uma condicao
print("produtos com preço abaixo de R$ 200")
produtos_baratos = df_produtos[df_produtos['Preco(R$)'] < 200]
print(produtos_baratos)

#calculando valor total de esoque para cada produto
df_produtos['Valor Total em Estoque'] = df_produtos['Preco(R$)'] * df_produtos['Quantidade em Estoque']

print("dataFrame com nova coluna de Valor Total")
print(df_produtos)
```

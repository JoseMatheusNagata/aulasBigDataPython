```python
import pandas as pd

#dataFrame da esquerda: funcionarios alocados e projetos
#note que o projeto C nao tem numbuem alocado aqui
funcionarios = pd.DataFrame({
    'id_funcionario': [101,102,103],
    'nome_funcionario': ['Joao', 'Maria', 'Pedro'],
    'id_projeto': ['A1','B2','A1']
})

#DataFrame da direita: lista completa de projetos
#o projeto D3 nao tem nunguem alocado ainda
projetos = pd.DataFrame({
    'id_projeto': ['A1','B2','C4','D3'],
    'nome_projeto': ['Desenvolvimento do AP','Migração de Dados','Analise de Mercado','Treinamento Interno']
})

print("dataFrame de Funcionarios: ")
print(funcionarios)
print('\n')
print("dataFrame de projetos")
print(projetos)

df_merge_right = pd.merge(funcionarios,projetos,on='id_projeto',how='right')
print('\n')
print("Data Frame com right Merge")
print(df_merge_right)
```
